Handling outliers

```python
### 5 numbers summary 
#1.Minimum value
#2.Q1-25percentile
#3.Median
#4.Q3-75percentile
#5.Maximum
```

```python
import numpy as  np
```

```python
lst_marks=[45,32,56,75,89,54,32,89,90,87,67,54,45,98,99,74,1000,1100]
```

```python
##[lower fence-->Higher fence]
Q1=np.percentile(lst_marks,[25])
```

```python
print(Q1)
```

```python
minimum,Q1,Q2,Q3,maximum=np.quantile(lst_marks,[0,0.25,0.50,0.75,1.0])
```

```python
maximum
```

```python
IQR=Q3-Q1
```

```python
print(IQR)
```

```python
lower_Fence=Q1-1.5*IQR
```

```python
print(lower_Fence)
```

```python
higher_Fence=Q3+1.5*IQR
```

```python
print(higher_Fence)
```

```python
outlier=[]
for i in lst_marks:
    if i>=lower_Fence and i<=higher_Fence:
        print("Not Outliers")
    else:
        outlier.append(i)
```
