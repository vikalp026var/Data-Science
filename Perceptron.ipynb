```python
import numpy as np 
import matplotlib.pyplot as plt
import joblib
import pandas as pd 
```

```python
class perceptron:
    def __init__(self,eta,epochs):
        self.weights=np.random.randn(3)*1e-4
        print(f"self.weights:{self.weights}")
        self.eta=eta
        self.epochs=epochs
    def activateFunctions(self,inputs,weights):
        z=np.dot(inputs,weights)
        return np.where(z>0,1,0)
    def fit(self,x,y):
        self.x=x
        self.y=y
        
        X_with_bias=np.c_[self.x,-np.ones((len(self.x),1))]
        print(f"X_with_bias:\n{X_with_bias}")
        
        for epoch in range(self.epochs):
            print(f"for epoch:{epoch}")
            y_hat=self.activateFunctions(X_with_bias,self.weights)
            print(f"prediction value:\n{y_hat}")
            error=self.y-y_hat
            print(f"error:\n{error}")
            self.weights=self.weights+self.eta*np.dot(X_with_bias.T,error)
            print(f"updated weights:\n{self.weights}")
            
    def predict(self,X):
        X_with_bias=np.c_[X,-np.ones((len(self.x),1))]
        return self.activateFunctions(X_with_bias,self.weights)
    
        
        
        
```

```python
data={"x1":[0,0,1,1],"x2":[0,1,0,1],"y":[0,0,0,1]}
AND=pd.DataFrame(data)
AND
```

```python
X=AND.drop("y",axis=1)
X
```

```python
y=AND['y']
y.to_frame()
```

```python
model=perceptron(eta=0.5,epochs=10)
model.fit(X,y)
```

```python
model.predict(X)
```

```python
model.weights
```
