```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip)
```

```python
tips.head()
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x=tips.total_bill,y=tips.tip,data=tips,hue='smoker')#hue gives color according to data 
```

```python
tips.head()
```

```python
sns.relplot(x=tips.total_bill,y=tips.tip,data=tips,style='size')#hue gives color according to data 
```

```python
tips['size'].value_counts()
```

```python
sns.relplot(x=tips.total_bill,y=tips.tip,data=tips,style='size',hue="time")#hue gives color according to data 
```

```python
sns.catplot(x='day',y='total_bill',data=tips)
```

```python
tips.head()
```

```python
sns.catplot(x='smoker',y='tip',data=tips)
```

```python
sns.jointplot(x=tips.total_bill,y=tips.tip)
```

```python
sns.pairplot(tips)
```
